# **Studi Kasus ANOVA: Efektivitas Masker dan Durasi Pemakaian terhadap Jumlah Partikel yang Tersaring**

Selama peningkatan kualitas udara menjadi perhatian global (misalnya akibat polusi atau kebakaran hutan), sebuah tim peneliti ingin menguji seberapa efektif berbagai jenis masker dalam menyaring partikel berbahaya (misalnya PM2.5) berdasarkan **jenis masker** dan **durasi pemakaian**.

Peneliti mengamati **jumlah partikel yang berhasil disaring** oleh berbagai jenis masker, setelah digunakan dalam durasi tertentu.

Mereka menggunakan:

* **Faktor A: Jenis Masker**

  * Masker Kain
  * Masker Bedah
  * Masker N95

* **Faktor B: Durasi Pemakaian**

  * 1 Jam
  * 4 Jam

Untuk masing-masing kombinasi, dilakukan pengujian laboratorium sebanyak 3 kali, dan tercatat jumlah partikel (dalam satuan standar) yang berhasil disaring.

Berikut data hasilnya:

```python
import pandas as pd

data = {
    "Masker": [
        "Kain", "Kain", "Kain", "Kain", "Kain", "Kain",
        "Bedah", "Bedah", "Bedah", "Bedah", "Bedah", "Bedah",
        "N95", "N95", "N95", "N95", "N95", "N95"
    ],
    "Durasi": [
        "1 Jam", "1 Jam", "1 Jam", "4 Jam", "4 Jam", "4 Jam",
        "1 Jam", "1 Jam", "1 Jam", "4 Jam", "4 Jam", "4 Jam",
        "1 Jam", "1 Jam", "1 Jam", "4 Jam", "4 Jam", "4 Jam"
    ],
    "Tersaring": [
        62, 64, 61,    # Kain - 1 jam
        59, 57, 56,    # Kain - 4 jam
        78, 80, 79,    # Bedah - 1 jam
        72, 74, 73,    # Bedah - 4 jam
        95, 96, 94,    # N95 - 1 jam
        90, 89, 91     # N95 - 4 jam
    ]
}

df = pd.DataFrame(data)
df
```

## **Pertanyaan:**

1. Apakah jenis masker memiliki pengaruh signifikan terhadap jumlah partikel yang tersaring?
2. Apakah durasi pemakaian memiliki pengaruh signifikan terhadap jumlah partikel yang tersaring?
3. Apakah terdapat interaksi antara jenis masker dan durasi pemakaian dalam memengaruhi jumlah partikel yang tersaring?

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Model ANOVA dengan interaksi
model = ols('Tersaring ~ C(Masker) + C(Durasi) + C(Masker):C(Durasi)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.barplot(x="Masker", y="Tersaring", hue="Durasi", data=df, errorbar=None)
plt.title("Jumlah Partikel Tersaring Berdasarkan Masker dan Durasi Pemakaian")
plt.ylabel("Partikel Tersaring")
plt.xlabel("Jenis Masker")
plt.grid(True)
plt.show()
```

## **Jawaban**

### 1. Apakah *jenis masker* berpengaruh terhadap jumlah partikel yang tersaring?

* F = 1108.19
* p-value = 2.438634e-14 (= 0.0000000000000243)

Iya, signifikan.
Karena p-value sangat kecil (< 0.05), kita menolak H. Artinya, jenis masker memiliki pengaruh signifikan terhadap jumlah partikel yang tersaring.

---

### 2. Apakah *durasi pemakaian* berpengaruh terhadap jumlah partikel yang tersaring?

* F = 88.62
* p-value = 6.85e-07 (= 0.000000685)

Iya, signifikan.
Karena p-value < 0.05, maka kita menolak H. Durasi pemakaian berpengaruh terhadap jumlah partikel yang tersaring.

---

### 3. Apakah ada *interaksi* antara jenis masker dan durasi pemakaian terhadap jumlah partikel yang tersaring?

* F = 0.35
* p-value = 0.7142396 (= 71.4%)

Tidak signifikan.
Karena p-value > 0.05, maka kita gagal menolak H. Artinya, tidak terdapat interaksi yang signifikan antara jenis masker dan durasi pemakaian.

---

### Kesimpulan Akhir:

* Jenis masker berpengaruh signifikan terhadap jumlah partikel yang tersaring
* Durasi pemakaian juga berpengaruh signifikan
* Namun tidak ada interaksi antara jenis masker dan durasi pemakaian, sehingga efek jenis masker tetap konsisten meskipun waktu pemakaian berbeda
